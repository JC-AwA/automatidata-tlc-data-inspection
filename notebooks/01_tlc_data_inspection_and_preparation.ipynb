```python
# 1. Convertir las columnas de fechas a formato datetime manejando formatos mixtos
# Al usar format='mixed', Pandas infiere inteligentemente los distintos formatos (MM/DD/YYYY vs MM-DD-YY)
date_cols = ['tpep_pickup_datetime', 'tpep_dropoff_datetime']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], format='mixed')
# 2. Filtrar 'RatecodeID' para mantener solo los valores en el rango esperado (1 a 6)
if 'RatecodeID' in df.columns:
    df = df[(df['RatecodeID'] >= 1) & (df['RatecodeID'] <= 6)]
# 3. Filtrar las columnas de montos para que todos los valores sean mayores o iguales a 0 (>= 0)
amount_cols = ['fare_amount', 'extra', 'mta_tax', 'improvement_surcharge', 'total_amount']
for col in amount_cols:
    if col in df.columns:
        df = df[df[col] >= 0]
# 4. Filtrar valores extremos en 'fare_amount' (como el 999.99)
if 'fare_amount' in df.columns:
    df = df[df['fare_amount'] < 450] # Descartamos valores de 450 o 999.99 a superiores
# Opcional: Mostrar un resumen para verificar que los rangos son correctos ahora
print("Limpieza completada. Filas restantes:", len(df))
display(df[amount_cols + ['RatecodeID']].describe())
```

### **Task 2c. Understand the data - Investigate the variables**

Sort and interpret the data table for two variables:`trip_distance` and `total_amount`.

**Answer the following three questions:**

**Question 1:** Sort your first variable (`trip_distance`) from maximum to minimum value, do the values seem normal?

**Question 2:** Sort by your second variable (`total_amount`), are any values unusual?

**Question 3:** Are the resulting rows similar for both sorts? Why or why not?

Q1: The values seem normal, yes.

Q2: So unusual values. Some 0 values in `trip_distance` match with some of the largest `total_amount` values.

Q3: No, the resulting rows are not exactly the same. Because while `trip_distance` and `total_amount` are naturally positively correlated (longer trips usually cost more), the `total_amount` is affected by several other factors besides just distance. The `total_amount` includes tips, tolls, and surcharges. For example, a trip might not be the absolute longest in distance, but if it included expensive bridge tolls, heavy traffic (time-based fare), or a very generous tip, it will rank much higher in the `total_amount` sorting than in the `trip_distance` sorting.

```python
# ==> ENTER YOUR CODE HERE

# Sort the data by trip distance from maximum to minimum value
df_trip_distance = df.sort_values(by='trip_distance', ascending=False)

# Opcional: Mostrar las primeras filas para verificar el ordenamiento
display(df_trip_distance.head(20))
```

```python
#==> ENTER YOUR CODE HERE

# Sort the data by total amount and print the top 20 values
df_distancexamount = df_trip_distance.sort_values(by='total_amount', ascending=False)
display(df_distancexamount.head(20))
# Si solo quieres ver esa columna especfica en lugar de todo el DataFrame, puedes usar:
# display(df_sorted_total['total_amount'].head(20))
```

```python
#==> ENTER YOUR CODE HERE

# What is the average tip for trips paid for with credit card?
# Filtrar por payment_type == 1 (Tarjeta de crdito) y calcular el promedio de 'tip_amount'
avg_tip_cc = df[df['payment_type'] == 1]['tip_amount'].mean()

print("Promedio de propina con tarjeta de crdito:", avg_tip_cc)


#==> ENTER YOUR CODE HERE

# What is the average tip for trips paid for with cash?
avg_tip_cash = df[df['payment_type'] == 2]['tip_amount'].mean()
print("Promedio de propina con efectivo:", avg_tip_cash)
```

```python
#==> ENTER YOUR CODE HERE

# How many times is each vendor ID represented in the data? 
vendor_counts = df['VendorID'].value_counts()
print(vendor_counts)
   
```

```python
#==> ENTER YOUR CODE HERE

# What is the mean total amount for each vendor?
mean_amount_by_vendor = df.groupby('VendorID')['total_amount'].mean()
print(mean_amount_by_vendor)
```

```python
#==> ENTER YOUR CODE HERE

# Filter the data for credit card payments only
credit_card_df = df[df['payment_type'] == 1]

# Opcional: Mostrar cuntas filas quedaron despus del filtro
print("Viajes pagados con tarjeta de crdito:", len(credit_card_df))
```

```python
#==> ENTER YOUR CODE HERE

# Calculate the average tip amount for each passenger count (credit card payments only)
avg_tip_by_passenger = credit_card_df.groupby('passenger_count')['tip_amount'].mean()
# Mostrar los resultados
print(avg_tip_by_passenger)
```
